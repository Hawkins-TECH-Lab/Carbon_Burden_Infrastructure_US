```python
df = gpd.read_parquet(tran_par_file)

if set(rows_keep).difference(df.columns):
    df = pd.merge(
        df,
        epa[['geoid20','noncom_vmt', 'com_vmt_pe', 'vmt_per_wo', 'vmt_tot_mi',
     'vmt_tot_ma', 'vmt_tot_av', 'ghg_per_wo', 'annual_ghg']], # Select only the columns you need from the second DF
        on='geoid20',
        how='left'
    )
df.to_parquet(tran_par_file)
df[rows_keep] = df.loc[:, rows_keep].replace(0, 1)
df.loc[:,rows_keep] = np.log(df.loc[:,rows_keep])
correlation_table = df.loc[:, cols].corr()
correlation_table = correlation_table.loc[rows_keep, cols_keep]
correlation_table.to_csv("/work/hawkins_lab/vulcan/results/correlation_table_transport.csv")
print(correlation_table)
```
