# Print results by sector and 5D

```python
trace = az.from_netcdf("/work/hawkins_lab/Mehrnoosh/vulcan-results/vulcan-results 02-22-2026/vulcan-elec_dens-pie.nc")
w = trace.posterior["w_mean"]  # dims: (chain, draw, w_mean_dim_0)
rhat_all = az.rhat(w)
print("Results for electricity density:")
print(rhat_all.mean())
print(rhat_all.quantile([0.1, 0.5, 0.9]))
```
